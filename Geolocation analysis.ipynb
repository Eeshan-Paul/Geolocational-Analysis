```python
from scipy import stats
outlier_list=[]
for i in new_data.columns[:]:
    q1=data[i].quantile(q=0.25)
    q3=data[i].quantile(q=0.75)
    L_outliers=q1-(1.5*(q3-q1))
    R_outliers=q3+(1.5*(q3-q1))
    outlier_list.extend(new_data[(new_data[i]<L_outliers) | (new_data[i]>R_outliers)].index)

print(outlier_list)
new_data.loc[list(set(outlier_list))]
new_data=new_data.drop(outlier_list,axis=0).reset_index(drop=True)
new_data.shape
```

```python
resta_others=result['response']['groups'][0]['items']
near_resta_others=pd.json_normalize(resta_others)
near_resta_others.head(5)
```

```python
restaurant=[]
others=[]
for lat,lng in zip(near_resta_others['venue.location.lat'],near_resta_others['venue.location.lng']):
    url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
    CLIENT_ID, 
    CLIENT_SECRET, 
    v, 
    lat,lng, # stands for latitude and longitude
    1000, 
    10)
    sub_result=requests.get(url).json()
    sub_resta_others=sub_result['response']['groups'][0]['items']
    near_sub_resta_others=pd.json_normalize(sub_resta_others)
    data=near_sub_resta_others['venue.categories']
    
    desires=[]
    for i in range(0,data.size):
        desires.append(data[i][0]['icon']['prefix'].find('food'))
    
    resta=0
    for i in desires:
        if i>1:
            resta+=1
    restaurant.append(resta)
    others.append(len(desires)-resta)

near_resta_others['restaurants']=restaurant
near_resta_others['others']=others
near_resta_others
```

```python
col=['venue.location.lat','venue.location.lng']
data=near_resta_others[col]
wcss=[]

for i in range(1,11):
    model=KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    model.fit(data)
    wcss.append(model.inertia_)

k = [i*100 for i in np.diff(wcss,2)].index(min([i*100 for i 
     in np.diff(wcss,2)]))
    
fig,ax=plt.subplots()
ax.plot(range(1,11),wcss)
ax.axvline(k, ls='--', color="red", label="k = "+str(k))
ax.set(title='The Elbow Method', xlabel='Number of clusters', 
       ylabel="Distortion")
ax.legend()
ax.grid(True)
plt.show()

```
